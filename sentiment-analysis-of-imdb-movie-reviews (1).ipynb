removing punctuation

```python
import string,time
```

```python
string.punctuation
```

```python
exclude = string.punctuation
```

```python
def remove_punc(text):
    for char in exclude:
        text = text.replace(char,'')
    return text
```

```python
text  = 'Hello! how are, you?'
```

```python
print(remove_punc(text))
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(text):
    # Tokenize the input text
    words = text.split()
    
    # Initialize a list to store non-stopwords
    new_words = []
    
    # Iterate through each word in the text
    for word in words:
        # Check if the word is not a stopword
        if word.lower() not in stopwords.words('english'):
            # If not a stopword, add it to the new list
            new_words.append(word)
    
    # Join the non-stopwords to form the new text
    new_text = " ".join(new_words)
    
    return new_text

# Apply the remove_stopwords function to the 'review' column
df['review_without_stopwords'] = df['review'].apply(remove_stopwords)
```
